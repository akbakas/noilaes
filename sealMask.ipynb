```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
import copy
```

```python
images = glob.glob('manualPick/*.JPG')
```

```python
def blur(img, k, s=1, method="Median"):
    gray = cv2.cvtColor(copy.copy(img), cv2.COLOR_BGR2GRAY)
    if method == "Gaussian":
        blurred = cv2.GaussianBlur(gray, k, s)
    elif method == 'Median':
        blurred = cv2.medianBlur(gray, k)
    else:
        blurred = None
    return blurred
```

```python
img = cv2.imread(images[346])
blurred = blur(img, 5)
```

```python
_, ret = cv2.threshold(blurred, 127, 255,cv2.ADAPTIVE_THRESH_MEAN_C)
# _, ret = cv2.threshold(blurred, 127, 255,cv2.THRESH_OTSU)
```

```python
plt.imshow(ret, cmap='gray')
plt.imshow(img, alpha=0.9)
```
